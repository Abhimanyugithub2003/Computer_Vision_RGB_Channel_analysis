```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image1=cv2.imread("/content/ChatGPT Image Mar 6, 2026, 04_20_11 PM.png")
plt.imshow(image1)
```

```python
image2=cv2.imread("/content/Radiant star in the darkness.png")
plt.imshow(image2)
```

```python

#and operator work as white and white portion become white as other is black

andimg=cv2.bitwise_and(image1,image2)
plt.imshow(andimg)

```

```python
#bitwise or act as if any one portion is white the portion become white

orimg=cv2.bitwise_or(image1,image2)
plt.imshow(orimg)
```

```python
not_img=cv2.bitwise_not(image1)
plt.imshow(not_img)
```

```python
not_img2=cv2.bitwise_not(image2)
plt.imshow(not_img2)
```

```python
xor_img=cv2.bitwise_xor(image1,image2)
plt.imshow(xor_img)
```
