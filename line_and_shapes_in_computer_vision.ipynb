```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img1=np.zeros(shape=(500,500,3),dtype=np.int32)
```

```python
img1.shape
```

```python
plt.imshow(img1)
```

```python
cv2.rectangle(img1,pt1=(300,50),pt2=(400,100),color=(0,0,255),thickness=2)
plt.imshow(img1)
```

```python
cv2.rectangle(img1,pt1=(300,300),pt2=(400,400),color=(0,255,0),thickness=3)
plt.imshow(img1)
```

```python
cv2.circle(img1,center=(50,50),radius=25,color=(255,0,0),thickness=3)
plt.imshow(img1)
```

```python
cv2.circle(img1,center=(250,250),radius=50,color=(255,0,0),thickness=-1)
plt.imshow(img1)
```

```python
cv2.line(img1,pt1=(500,500),pt2=(0,0),color=(255,255,255),thickness=3)
plt.imshow(img1)
```

```python
#polygon and text in images
img2=np.zeros(shape=(500,500,3),dtype=np.int32)
plt.imshow(img2)
```

```python
font=cv2.FONT_HERSHEY_COMPLEX
cv2.putText(img2,text='ABHIMANYU',org=(10,450),fontFace=font,fontScale=3,color=(255,255,255),thickness=2)
plt.imshow(img2)
```

```python
img3=np.zeros(shape=(500,500,3),dtype=np.int32)
vertices=np.array([[50,50],[400,50],[400,300],[100,300]],np.int32)
cv2.polylines(img3,[vertices],isClosed=True,color=(255,0,0),thickness=3)
plt.imshow(img3)
```
